```python
import numpy as np
from scipy.linalg import expm, sinm, cosm
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import math
from scipy.special import iv
```

```python
def theoretical_acf(rho, upper_bound):
    acf = []
    acf_denom = iv(0, 2.0 * rho)

    for h in range(0,upper_bound):
        acf_num = iv(h, 2.0*rho)
        
        
        acf.append( (acf_num/acf_denom).real)
        
    return acf
```

```python
pd.DataFrame(theoretical_acf(3.0,15)).plot()
```

```python
def mets_forecast(rho,Y , max_length):
    if len(Y) < max_length:
        np.append( np.zeros(max_length - len(Y)), Y)
        
    Y = np.flip(Y[-max_length:])
    return -1.0*np.sum([y*rho**h/math.factorial(h) for y,h in zip(Y,range(1,max_length-1))])
```

```python
ts = np.random.normal(0,1,10)

for k in range(1,10):
    ts = np.append(ts, mets_forecast(1.0,ts,10)) 
    
pd.DataFrame(ts).plot()
```

```python
forecast_df = pd.DataFrame( columns = range(0,28))

for rho in np.linspace(0,-1,20):
    rho_ts = ts
#     print(rho_ts)
#     print()
    
    for k in range(1,10):
        new_forecast = mets_forecast(rho,rho_ts,15)
#         print(new_forecast)
        rho_ts = np.append(rho_ts, new_forecast) 
    forecast_df.loc[rho] = rho_ts
```
